```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
pd.set_option('display.max_columns', None)
oly_gdp_hdi = pd.read_csv("oly_gdp_hdi.csv") # table with medal counts, medal proportions, country GDP, country HDI
```

```python
# home field advantage at summer olympics
summer_prop = oly_gdp_hdi.filter(['2020_mp', '2016_mp', '2012_mp', '2008_mp', '2004_mp', '2000_mp', '1996_mp', 
                                  '1992_mp', '1988_mp'])
summer_year = [2020, 2016, 2012, 2008, 2004, 2000, 1996, 1992, 1988] 
summerhost_mp = summer_prop.loc[[10, 25, 34, 65, 66, 86, 159, 160, 179], :]

# line graph
plt.subplots(1, figsize = (10, 8))
plt.margins(x=0, y=0)
plt.plot(summer_year, summer_prop.loc[10].transpose()*100, label = "Australia")
plt.plot(summer_year, summer_prop.loc[25].transpose()*100, label = "Brazil")
plt.plot(summer_year, summer_prop.loc[34].transpose()*100, label = "China")
plt.plot(summer_year, summer_prop.loc[65].transpose()*100, label = "Great Britain")
plt.plot(summer_year, summer_prop.loc[66].transpose()*100, label = "Greece")
plt.plot(summer_year, summer_prop.loc[86].transpose()*100, label = "Japan")
plt.plot(summer_year, summer_prop.loc[159].transpose()*100, label = "South Korea")
plt.plot(summer_year, summer_prop.loc[160].transpose()*100, label = "Spain")
plt.plot(summer_year, summer_prop.loc[179].transpose()*100, label = "United States")
plt.plot(2020, 5.2974, 'ko')
plt.plot(2016, 1.9608, 'ko')
plt.plot(2012, 6.8063, 'ko')
plt.plot(2008, 10.4822, 'ko')
plt.plot(2004, 1.7316, 'ko')
plt.plot(2000, 6.2907, 'ko')
plt.plot(1996, 12.0525, 'ko')
plt.plot(1992, 2.7446, 'ko')
plt.plot(1988, 3.88694, 'ko')
plt.annotate("Tokyo 2020", (2020, 5.2974), xytext = (2017, 5.4))
plt.annotate("Rio 2016", (2016, 1.9608), xytext = (2016, 2.1))
plt.annotate("London 2012", (2012, 6.8063), xytext = (2012, 7.0))
plt.annotate("Beijing 2008", (2008, 10.4822), xytext = (2008, 10.6))
plt.annotate("Athens 2004", (2004, 1.7316), xytext = (2004, 1.9))
plt.annotate("Sydney 2000", (2000, 6.2907), xytext = (2000, 6.4))
plt.annotate("Atlanta 1996", (1996, 12.0525), xytext = (1996, 12.2))
plt.annotate("Barcelona 1992", (1992, 2.7446), xytext = (1992, 2.9))
plt.annotate("Seoul 1988", (1988, 3.88694), xytext = (1988, 4.0))
plt.xticks(summer_year)
plt.xlim([1987, 2021])
plt.ylim([0, 14.5])
plt.xlabel('Year')
plt.ylabel('Percent of Medals Won')
plt.title('Home Field Advantage at the Summer Olympic Games')
plt.legend(loc = 'center left', bbox_to_anchor = (1, 0.5))
plt.savefig('home_summer.png', bbox_inches = 'tight')
```
