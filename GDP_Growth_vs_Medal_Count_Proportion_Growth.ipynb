```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
pd.set_option('display.max_columns', None)

oly = pd.read_csv("oly_gdp_hdi.csv")
oly[oly['country']=='China']
```

```python
columns = ['2020_mp', '1992_mp', '2020_gdp', '1992_gdp', '2018_hdi', '1992_hdi', 'country']
```

```python
aus = oly[oly['country'] == 'Australia'][columns]
can = oly[oly['country'] == 'Canada'][columns]
fra = oly[oly['country'] == 'France'][columns]
ita = oly[oly['country'] == 'Italy'][columns]
jap = oly[oly['country'] == 'Japan'][columns]
nor = oly[oly['country'] == 'Norway'][columns]
rus = oly[oly['country'] == 'Russia'][columns]
sot = oly[oly['country'] == 'South Korea'][columns]
ger = oly[oly['country'] == 'Germany'][columns]
ukr = oly[oly['country'] == 'Ukraine'][columns]
hun = oly[oly['country'] == 'Hungary'][columns]
net = oly[oly['country'] == 'Netherlands'][columns]
bra = oly[oly['country'] == 'Brazil'][columns]
rom = oly[oly['country'] == 'Romania'][columns]
pol = oly[oly['country'] == 'Poland'][columns]
fin = oly[oly['country'] == 'Finland'][columns]
chin = oly[oly['country'] == 'China'][columns]
swi = oly[oly['country'] == 'Switzerland'][columns]
por = oly[oly['country'] == 'Portugal'][columns]
sov = oly[oly['country'] == 'Soviet Union'][columns]
nig = oly[oly['country'] == 'Nigeria'][columns]
mex = oly[oly['country'] == 'Mexico'][columns]
gre = oly[oly['country'] == 'Greece'][columns]
chil = oly[oly['country'] == 'Chile'][columns]
cro = oly[oly['country'] == 'Croatia'][columns]
egy = oly[oly['country'] == 'Egypt'][columns]
ire = oly[oly['country'] == 'Ireland'][columns]
moro = oly[oly['country'] == 'Morocco'][columns]
spa = oly[oly['country'] == 'Spain'][columns]
gua = oly[oly['country'] == 'Guatemala'][columns]
colo = oly[oly['country'] == 'Colombia'][columns]

us = oly[oly['country'] == 'United States'][columns]
chin = oly[oly['country'] == 'China'][columns]

```

```python
oly_short = pd.concat([aus, can, fra, ita, jap, nor, rus, sot, nor, fin, pol, rom, net, hun, ukr, ger, nig, sov, por, swi, colo, gua, spa, moro, ire, egy, cro, chil, gre, mex])
```

```python
oly_short
```

```python
oly_short['gdp_growth_factor'] = oly_short['2020_gdp']-oly_short['1992_gdp']
```

```python
oly_short['mp_growth_factor'] = oly_short['2020_mp']/oly_short['1992_mp']
```

```python
# plt.xlabel("GDP Growth Factor")
# plt.ylabel("Medal Count Growth Factor")
# plt.title("GDP Growth vs Medal Count Growth")
# plt.scatter(gdp_factor, mp)
```

```python
fig, ax = plt.subplots(1, figsize=(10,10))
ax.scatter(oly_short["gdp_growth_factor"], oly_short["mp_growth_factor"])
plt.xlabel("Total GDP Growth")
plt.ylabel("Medal Count Proportion Growth Factor")
plt.title("GDP Growth vs Medal Count Growth")
oly_short[['gdp_growth_factor','mp_growth_factor','country']].apply(lambda x: ax.text(*x),axis=1)
```
