# 2) Removing Olympic Games before 1986

https://medium.com/fhsaplang/why-1986-changed-the-olympics-forever-b4aefb740199

```python
oly_results["year"] = oly_results["slug_game"].str.split("-").str.get(-1)
oly_results["year"].unique()
#I'm not sure why 1990 and 1986
```

```python
oly_results["year"] = oly_results["year"].astype(int)
oly_results = oly_results[oly_results.year >= 1986]
```

# 3) Adding data from the olympic_hosts.csv file

```python
oly_hosts = pd.read_csv("olympic_hosts.csv")
oly_hosts.head()
```

```python
oly_hosts = oly_hosts.rename(columns={"game_slug":"slug_game"})
oly_hosts.head()
```

```python
oly = pd.merge(oly_results, oly_hosts, how="left", on= "slug_game")
```

```python
print(oly_results.shape)
print(oly.shape)
#so we keep all the rows, but add informational colums for oly_hosts
```

```python
oly.head()
```

# 4) Recoding some country names

```python
oly["country_name"].unique()
```

```python
#
def recode_countries(name):
    if name == "ROC":
        return "Russia"
    elif name == "Olympic Athletes from Russia":
        return "Russia"
    elif name == "Soviet Union":
        return "Russia"
    elif name == "Russian Federation":
        return "Russia"
    elif name == "Federal Republic of Germany":
        return "Germany"
    elif name == "German Democratic Republic (Germany)":
        return "Germany"
    elif name == "Republic of Korea":
        return "South Korea"
    elif name == "Democratic People's Republic of Korea":
        return "South Korea"
    elif name == "Korea Team":
        return "South Korea"
    elif name == "United States of America":
        return "United States"
    elif name == "People's Republic of China":
        return "China"
    else:
        return name
```

```python
oly["country_test"] = oly["country_name"].apply(recode_countries)
oly.head()
```

```python
oly = oly.rename(columns={"country_test":"country"})
oly.head()
```

# 5) "Unified Team" Country

https://en.wikipedia.org/wiki/Unified_Team_at_the_1992_Summer_Olympics

```python
oly = oly[oly["country"] != "Unified Team"]
```

# 6) True/False Column for if they won a medal

```python
set(oly.medal_type)
```

```python
recode = {
    np.nan : False,
    "BRONZE" : True,
    "SILVER" : True,
    "GOLD" : True
}

oly["medal"] = oly["medal_type"].map(recode)
```

# 7) Saving cleaned datatable as a csv file

```python
oly.to_csv('oly.csv')
```
